```python
sediment_soil_GO_ratio = np.log2(data.loc['root:Mixed:Sediment:Sediment:Sediment']/data.loc['root:Mixed:soil:soil:soil'])
```

```python
sediment_soil_GO_ratio
```

```python
ratio_df = pd.DataFrame(sediment_soil_GO_ratio, columns=['ratio'])
```

```python
ratio_df
```

```python
print(ratio_df.min())
print(ratio_df.max())
```

```python
ratio_df_filter = ratio_df.loc[(ratio_df['ratio'] > 10) | (ratio_df['ratio'] < -10)]
ratio_df_filter = ratio_df_filter.sort_values(by=['ratio'])
ratio_df_filter.shape
```

```python
x_pos = [i for i, _ in enumerate(ratio_df_filter.index)]

plt.bar(x_pos, ratio_df_filter.iloc[:,0], color='green')
plt.xlabel("GO terms")
plt.ylabel("log ratio")
plt.title("GO term over/under representation: sediment vs soil")

plt.xticks(x_pos, ratio_df_filter.index)

plt.show()
```

```python
df_shape = data.shape
#for i in range(0,df_shape[0]:
#    for j in range(0,df_shape[1]:                   
                   

row_labels = []

for i in range(0,df_shape[0]):
    for j in range(0,df_shape[0]):
        if(i != j):
              row_labels.append(data.index[i]+"__vs__"+data.index[j])

ratio_df_all = pd.DataFrame(index=row_labels, columns=data.columns)  
    
            
rows_list = []
count = 0
for i in range(0,df_shape[0]):
    for j in range(0,df_shape[0]):
        if(i != j):
            curdata = np.log2(data.iloc[i]/data.iloc[j])
            #print(curdata)
            ratio_df_all.iloc[count] = curdata
            #print(curdata.shape)            
            #rows_list.append(curdata)
            count = count + 1
            #print(len(rows_list))
```
